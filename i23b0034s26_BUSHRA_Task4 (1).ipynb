
Practice problem 2.2

Write a Pandas program to get the day of month, day of year, week number and day of week from a given series of date strings.
 

Output:

Original Series:

0         01 Jan 2015

1          10-02-2016

2            20180307

3          2014/05/06

4          2016-04-12

5    2019-04-06T11:20

dtype: object

 

Day of month:

[1, 2, 7, 6, 12, 6]

Day of year:

[1, 276, 66, 126, 103, 96]

Week number:

[1, 39, 10, 19, 15, 14]

Day of week:

['Thursday', 'Sunday', 'Wednesday', 'Tuesday', 'Tuesday', 'Saturday']

```python
date = pd.Series(['01 Jan 2015', '10-02-2016', '20180307', '2014/05/06', '2016-04-12', '2019-04-06T11:20'])

print("Original Series:")

date
```

```python
date = pd.to_datetime(date, format='mixed')
```

```python
month = date.dt.day.tolist()

print("Day of month:")

month
```

```python
year = date.dt.dayofyear.tolist()
print("Day of year:")
year

```

```python
week = date.dt.isocalendar().week.tolist()

print("Week number:")
week
```

```python
day = date.dt.strftime('%A').tolist()
print("Day of week:")
day
```
